```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
PATH = './chromedriver'
driver = webdriver.Chrome(PATH)
chromeOptions = webdriver.ChromeOptions()
```

```python
driver.get("https://scholar.google.com/")
```

```python
#search page
search_box=driver.find_element_by_css_selector("#gs_hdr_tsi")
search_box.send_keys("Thammasat University\n")
```

```python
#click thammasat University
tu=driver.find_element_by_css_selector("#gs_res_ccl_mid > div:nth-child(2) > div > a")
tu.click()
```

```python
from selenium.common.exceptions import NoSuchElementException
df2 = pd.DataFrame()
for i in range(30):
    data = driver.find_elements_by_css_selector('#gsc_sa_ccl > div > div > div') 
    for i in range(len(data)):
        #click page
        data = driver.find_elements_by_css_selector('#gsc_sa_ccl > div > div > div')
        author=data[i].find_element_by_css_selector('#gsc_sa_ccl > div > div > div > h3 > a')
        author.click()
        time.sleep(2)
        while True:
            driver.find_element_by_css_selector("#gsc_bpf_more").click()
            time.sleep(2)
            if not driver.find_element_by_css_selector("#gsc_bpf_more").is_enabled():
                    break
                    
            for i in driver.find_elements(By.CSS_SELECTOR, "td.gsc_a_t"):
                webtittle=i.find_element_by_css_selector('a')
                webtittle.click()
                time.sleep(1.5)
                tittle=i.find_element_by_xpath("//*[@id='gsc_vcd_title']/a")
                t=tittle.text
                authors=driver.find_element_by_css_selector("#gsc_vcd_table > div:nth-child(1) > div.gsc_vcd_value")
                a=authors.text
                publication_date=driver.find_element_by_css_selector("#gsc_vcd_table > div:nth-child(2) > div.gsc_vcd_value")
                p=publication_date.text
                try:
                    description=driver.find_element_by_css_selector("#gsc_vcd_descr > div")
                    d=description.text  
                except NoSuchElementException:
                    pass
                #close page
                df2 = df2.append(
                {
                    'tittle':t,
                    'authors':a,
                    'publication_date':p,
                    'description':d
                }, ignore_index=True
                )
                driver.find_element_by_css_selector("#gs_md_cita-d-x").click()
                for i in driver.find_elements(By.CSS_SELECTOR, "td.gsc_a_c"):
                    cite_by=i.find_element_by_css_selector("a")
                    c=cite_by.text
                    df2['cite_by'] = c
                    print(c)
                               
        time.sleep(5)    
        driver.back()
        time.sleep(3)
    
    next_page=driver.find_element_by_css_selector('#gsc_authors_bottom_pag > div > button.gs_btnPR.gs_in_ib.gs_btn_half.gs_btn_lsb.gs_btn_srt.gsc_pgn_pnx')
    next_page.click()
    df2
    
df2
    
```
