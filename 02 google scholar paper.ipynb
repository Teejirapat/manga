```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
PATH = './chromedriver'
driver = webdriver.Chrome(PATH)
chromeOptions = webdriver.ChromeOptions()
```

```python
driver.get("https://scholar.google.com/")
```

```python
#search page
search_box=driver.find_element_by_css_selector("#gs_hdr_tsi")
search_box.send_keys("Thammasat University\n")
```

```python
#click thammasat University
tu=driver.find_element_by_css_selector("#gs_res_ccl_mid > div:nth-child(2) > div > a")
tu.click()
```

```python
from selenium.common.exceptions import NoSuchElementException
df2 = pd.DataFrame()
#loop for next page
for i in range(30):
    data = driver.find_elements_by_css_selector('#gsc_sa_ccl > div > div > div')
    #loop for click author
    for i in range(len(data)):
        #click page
        data = driver.find_elements_by_css_selector('#gsc_sa_ccl > div > div > div')
        author=data[i].find_element_by_css_selector('#gsc_sa_ccl > div > div > div > h3 > a')
        author.click()
        time.sleep(3)
        #loop for show all list
        while True:
            driver.find_element_by_css_selector("#gsc_bpf_more").click()
            time.sleep(3)
            if not driver.find_element_by_css_selector("#gsc_bpf_more").is_enabled():
                break
        #loop for show all information and put information in the pandas(df2)            
        for i in driver.find_elements(By.CSS_SELECTOR, "td.gsc_a_t"):
                webtittle=i.find_element_by_css_selector('#gsc_a_b > tr > td> a')
                webtittle.click()
                time.sleep(3)
                try:
                    tittle=i.find_element_by_xpath("//*[@id='gsc_vcd_title']/a")
                    t=tittle.text
                except NoSuchElementException:
                    t=""
                    pass
                try:
                    authors=driver.find_element_by_css_selector("#gsc_vcd_table > div > div.gsc_vcd_value")
                    a=authors.text
                except NoSuchElementException:
                    a=""
                    pass
                try:
                    publication_date=driver.find_element_by_css_selector("#gsc_vcd_table > div:nth-child(2) > div.gsc_vcd_value")
                    p=publication_date.text
                except NoSuchElementException:
                    p=""
                    pass
                try:
                    description=driver.find_element_by_css_selector("#gsc_vcd_descr > div")
                    d=description.text
                except NoSuchElementException:
                    d=""
                    pass
                try:
                    cite_by=driver.find_element_by_css_selector('#gsc_vcd_table > div> div> div > a').text
                    c=cite_by.split("อ้างโดย")
                    cite=c[1]
                except NoSuchElementException:
                    c=""
                    pass
                df2 = df2.append(
                {
                    'tittle':t,
                    'authors':a,
                    'publication_date':p,
                    'description':d,
                    'cite_by':cite
                }, ignore_index=True
                )
                #close page
                driver.find_element_by_css_selector("#gs_md_cita-d-x").click()
                time.sleep(2.5)
                df2
                               
        time.sleep(4)    
        driver.back()
        time.sleep(3)
    
    next_page=driver.find_element_by_css_selector('#gsc_authors_bottom_pag > div > button.gs_btnPR.gs_in_ib.gs_btn_half.gs_btn_lsb.gs_btn_srt.gsc_pgn_pnx')
    next_page.click()
```
