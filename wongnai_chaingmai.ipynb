```python
from selenium import webdriver
import time
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
```

```python
PATH = './chromedriver'
driver = webdriver.Chrome(PATH)
driver.get("https://www.wongnai.com/attractions?regions=373")
```

```python
df = pd.DataFrame()
length = len(driver.find_elements(By.CSS_SELECTOR, '#body > div> div > div.row > div.col-md-6 > div.sc-195dyzv-0.fkmkTf > div > div > a'))
print(length)
for i in range(length):
    try:
        data = driver.find_elements_by_css_selector('#body > div> div > div.row > div.col-md-6 > div.sc-195dyzv-0.fkmkTf > div> div > div > div.sc-AxiKw.sc-1nkxss0-2.fWKzcG > a')
        name = data[i].find_element_by_css_selector('h2')
        NAME = name.text
        print(NAME)
        name.click()
        time.sleep(2)
    except NoSuchElementException:
        name=" "
        pass
    address = driver.find_element_by_css_selector('#body > div.content.noSidePadding > div > div.iowx8b-3.fjJrOy.col-md-8 > div:nth-child(3) > div.sc-1xjzh3o-0.gFLGey > div > div.sc-1xjzh3o-2.bjDa-do > span > div > a > div > span.sc-18y57ed-3.gnUBUT > span > div > div')
    ADDRESS = address.text
    print(ADDRESS)
    try:
        phone = driver.find_element_by_css_selector('#body > div.content.noSidePadding > div > div.iowx8b-3.fjJrOy.col-md-8 > div:nth-child(3) > div.sc-1dcw2xg-0.gpXucT.iowx8b-4.gHJbkx > div > div:nth-child(1) > a > div > span.sc-18y57ed-3.gnUBUT > span > span')
        PHONE = phone.text.split(":")[-1]
        print(PHONE)
    except NoSuchElementException:
        PHONE =" "
        pass
    df = df.append(
                {
                    'The attraction':NAME,
                    'Phone':PHONE,
                    'address':ADDRESS
                }, ignore_index=True
                )
    driver.back()
    time.sleep(2)
    
```

```python

# save data
df.to_csv('chaingmai.csv')
#show data
df
```
